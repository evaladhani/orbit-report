## Prepare Your Dataframe

Import any libraries you need and create a dataframe.

```python
# Import potentially useful libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np

# Read dataset
df = pd.read_csv('top10s.csv', encoding = "ISO-8859-1")

# Print data set
df

# Run decriptive analysis
df.describe()
```

## Find Correlations

Use either pairplot or a heatmap or both to find the two variables with the strongest correlation in your dataset.

```python
#df = df.drop(['Unnamed: 0'], axis = 1)
df = df.rename(columns={"bpm": "Beats Per Minute", "nrgy": "Energy", "dnce": "Danceability", "dB":"loudness", 
                        "live":"Liveness", "val":"Valence", "dur":"Length", "acous":"Acousticness", 
                        "spch":"Speechiness", "pop":"Popularity"})

df.corr()
plt.subplots(figsize=(13,10))
sb.heatmap(df.corr())
```

## Create Your Model

Use the two columns with the strongest correlation to create and train your model. Make sure to print out the summary and plot the column values and the line produced by the model.

```python
df[['Acousticness', 'Energy']].corr()
```

```python
fig, ax = plt.subplots(figsize=(25,30))

ax.set_ylabel('Energy')

_ = df[['Acousticness', 'Energy']].boxplot(by='Energy', figsize=(13,8), vert=False, sym='b.', ax=ax)
```

```python
plt.figure(figsize=(14,8)) # set the size of the graph
_ = sb.regplot(data=df, x='Acousticness', y='Energy')
```

## Error Terms

Finally, plot your error terms!

```python
# Create tabular correlation matrix

corr = df.corr()
_, ax = plt.subplots(figsize=(13,10))

# Graph correlation matrix

_ = sb.heatmap(corr, ax=ax,
               xticklabels=corr.columns.values,
               yticklabels=corr.columns.values,
               cmap='coolwarm')
```
